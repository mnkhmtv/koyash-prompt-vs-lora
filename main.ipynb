```python
import pandas as pd
```

```python
df = pd.read_csv("metrics_results.csv")
```

```python
df = df.dropna()
```

```python
df.to_csv("metrics_results.csv")
```

```python
perplexity = df[df['model_name'] == 'BaselineKoyashLLM: qwen2.5:7b']['perplexity']
print(f"Mean perplexity: {perplexity.mean()}, Std perplexity: {perplexity.std()}")
rouge_l = df[df['model_name'] == 'BaselineKoyashLLM: qwen2.5:7b']['rouge_l']
print(f"Mean ROUGE-L: {rouge_l.mean()}, Std ROUGE-L: {rouge_l.std()}")
bert_s = df[df['model_name'] == 'BaselineKoyashLLM: qwen2.5:7b']['bert_score']
print(f"Mean BERTScore: {bert_s.mean()}, Std BERTScore: {bert_s.std()}")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

for ax, col, title, color in zip(
    axes,
    ["perplexity", "rouge_l", "bert_score"],
    ["Perplexity (↓ лучше)", "ROUGE-L (↑ лучше)", "BERTScore (↑ лучше)"],
    ["steelblue", "seagreen", "darkorange"],
):
    for model_name, group in df.groupby("model_name"):
        ax.hist(group[col], bins=15, alpha=0.7, label=model_name, color=color)
        ax.axvline(group[col].mean(), color="red", linestyle="--", linewidth=1.5, label=f"mean={group[col].mean():.3f}")
    ax.set_title(title)
    ax.set_xlabel(col)
    ax.set_ylabel("count")
    ax.legend(fontsize=7)

plt.tight_layout()
plt.show()
```
